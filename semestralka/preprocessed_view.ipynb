```python
import pandas as pd
```

```python
df1 = pd.read_csv('processed/playstore.csv')
df2 = pd.read_csv('processed/playstore_reviews.csv')
df3 = pd.read_csv('processed/cities.csv')
```

```python
df1.head()
```

```python
df1.App = df1.App.str.replace('"', '', regex=False).str.replace(',', '', regex=False).str.replace("''", '', regex=False)
df1.Category = df1.Category.str.replace('"', '', regex=False).str.replace(',', '', regex=False).str.replace("''", '', regex=False)
df1.Genres = df1.Genres.str.replace('"', '', regex=False).str.replace(',', '', regex=False).str.replace("''", '', regex=False)
df1["Android Ver"] = df1["Android Ver"].str.replace('"', '', regex=False).str.replace(',', '', regex=False).str.replace("''", '', regex=False)

df1.rename(index=str, columns={"Android Ver": "Android_Ver", "Content Rating": "Content_Rating", "Last Updated":"Last_Updated", "Current Ver":"Current_Ver"}, inplace=True)
#Content Rating    64654 non-null object
#Genres            64660 non-null object
#Last Updated      64660 non-null object
#Current Ver       64602 non-null object
#Android Ver       64636 non-null object
```

```python
df1.head()
```

```python
df1.Android_Ver.unique()
df1.Current_Ver = df1.Genres.str.replace('"', '', regex=False).str.replace(',', '', regex=False).str.replace("''", '', regex=False)
```

```python
df1.Last_Updated = df1.Last_Updated.str.replace(',', '/', regex=False).str.replace(' ', '/', regex=False)
```

```python
df1.Last_Updated = df1.Last_Updated.str.replace('//', '/', regex=False)
```

```python
df1.head()
```

```python
df2.head()
```

```python
df2.App = df2.App.str.replace('"', '', regex=False).str.replace(',', '', regex=False).str.replace("''", '', regex=False)
```

```python
df2.Translated_Review = df2.Translated_Review.str.replace('"', '', regex=False).str.replace(',', '', regex=False).str.replace("''", '', regex=False)
```

```python
df3.Region= df3.Region.str.replace('"', '', regex=False).str.replace(',', '', regex=False).str.replace("''", '', regex=False)
```

```python
df3.head()
```

```python
df1.to_csv('processed/playstore_2.csv', index=False)
df2.to_csv('processed/playstore_reviews_2.csv', index=False)
df3.to_csv('processed/cities_2.csv', index=False)
```
